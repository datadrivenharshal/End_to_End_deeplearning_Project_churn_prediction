```python
data['Total Charges']=pd.to_numeric(data['Total Charges'], errors='coerce')
data.dtypes

```

```python
#divide data into dependent and independent features

X=data.drop('Churn Value',axis=1)
y=data['Churn Value']


#train and test split

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)

#scale the features

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)

```

```python
##Ann Implementation
```

import tensorflow as tf

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
(X_train.shape[1],)
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)), #first hiddenlayer
    Dense(32,activation='relu'), #HL2
    Dense(1,activation='sigmoid'), #Output layer
    Dropout(0.2),
]
)
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
#compile the model
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
## Set up the tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

log_dir="log/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback=TensorBoard(log_dir,histogram_freq=1)
```

```python
#setup EarlyStopping

early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(X_train,y_train, validation_data=(X_test,y_test),epochs=100,callbacks=[tensorboard_callback,early_stopping_callback],batch_size=20)
```

```python
model.save('model.h5')
```
