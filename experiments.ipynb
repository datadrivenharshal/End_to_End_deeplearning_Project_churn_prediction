```python
#Encode categorical variable
label_encoder_gender=LabelEncoder()
label_encoder_senior_citizen=LabelEncoder()
label_encoder_partner=LabelEncoder()
label_encoder_dependents=LabelEncoder()
label_encoder_phoneservice=LabelEncoder()
label_encoder_multiple_lines=LabelEncoder()
label_encoder_online_security=LabelEncoder()
label_encoder_online_backup=LabelEncoder()
label_encoder_device_protection=LabelEncoder()
label_encoder_tech_support=LabelEncoder()
label_encoder_streaming_tv=LabelEncoder()
label_encoder_streaming_movies=LabelEncoder()


data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data['Senior Citizen']=label_encoder_senior_citizen.fit_transform(data['Senior Citizen'])
data['Partner']=label_encoder_partner.fit_transform(data['Partner'])
data['Dependents']=label_encoder_dependents.fit_transform(data['Dependents'])
data['Phone Service']=label_encoder_phoneservice.fit_transform(data['Phone Service'])
data['Multiple Lines']=label_encoder_multiple_lines.fit_transform(data['Multiple Lines'])
data['Online Backup']=label_encoder_online_backup.fit_transform(data['Online Backup'])

data['Device Protection']=label_encoder_device_protection.fit_transform(data['Device Protection'])

data['Online Security']=label_encoder_online_security.fit_transform(data['Online Security'])

data['Tech Support']=label_encoder_tech_support.fit_transform(data['Tech Support'])

data['Streaming TV']=label_encoder_streaming_tv.fit_transform(data['Streaming TV'])

data['Streaming Movies']=label_encoder_streaming_movies.fit_transform(data['Streaming Movies'])


```
